# ***Importing the dependencies***


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset
credit_card_data = pd.read_csv('/content/credit_card.csv')
```

```python
#first 5 rows of dataset
credit_card_data.head()
```

```python
#last 5 rows of dataset
credit_card_data.tail()
```

```python
#info about the data
credit_card_data.info()
```

```python
#check for missing valuse in each column
credit_card_data.isnull().sum()
```

```python
#distribution of legit transaction and fraudulent transactions
credit_card_data['Class'].value_counts()
```

# **This dataset is highly unbalanced**

0->normal transaction

1->fraudlent transaction

```python
#separating data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#getting some statistical measures
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

# **Under-sampling**

Buildig a simple dataset containing similar distribution of normal (legit) and
fraudulent transactions

number of fraudulent transactions --> 2

```python
legit_sample = legit.sample(n=2)
```

# **Concatenating two Dataframe**

```python
new_dataset = pd.concat([legit_sample, fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

#split the data intro features & targets






```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

# **Split the data into Training Data and Testing Data**

```python
X_train, X_test ,Y_train ,Y_test = train_test_split(X, Y, test_size=0.3, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# **Model** **Training**

# *logistic Regression Model*

```python
model = LogisticRegression()
```

```python
# training the logistic regression model with training data
model.fit(X_train, Y_train)
```

# **Model** **Evaluation**

# *Accuracy Score*

```python
#Accuracy on Training Data
X_train_predication = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_predication,Y_train)
```

```python
print('Accuracy on Training Data:' , training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test data:', test_data_accuracy)
```
